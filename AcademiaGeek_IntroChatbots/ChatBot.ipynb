### Importando a biblioteca

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download("punkt")
```

###Definindo a minha sentença

```python
texto = "Eu Lais estou aprendendo introdução ao desenvolvimento de chatbots"
```

###Separando o texto em sentenças

```python
print(":::::::SENTENÇAS:::::::")

#Função para separar o texto em sentenças 

print(sent_tokenize(texto))
```

###Tokenização da sentença

```python
print("\n::::::TOKENIZAÇÃO::::::")

#Função para tokenizar uma sentença

print(word_tokenize(texto))
```
